# Notebook setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

Function `stats` will return 5-tuple (min, max, mean, median, std)

```python
def stats(file):
    with open(file, 'r') as ifile:
        l = [float(line) for line in ifile.readlines()]
    array = np.array(l)
    return (np.min(array),
            np.max(array),
            np.mean(array),
            np.median(array),
            np.std(array))
```

Function `plot` plots 2 histograms:
1. Original data
1. Filtered data

_Filtered_ data doesn't contain outliers. See http://en.wikipedia.org/wiki/68%E2%80%9395%E2%80%9399.7_rule

```python
def plot(file):
    with open(file, 'r') as ifile:
        l = [float(line) for line in ifile.readlines()]

    array = np.array(l)
    criteria = np.median(array) + 3 * np.std(array)
    array_filtered = np.array([el for el in l if el < criteria])
    plt.figure().set_size_inches(16, 5)
    plt.title(file)
    plt.subplot(1,2,1)
    _ = plt.hist(array, bins=75)
    plt.subplot(1,2,2)
    _ = plt.hist(array_filtered, bins=75)
```

# Experimenting

Main difference in these experiments is not taking into account creating/closing SSL connection. Just measure how much time single `HEAD` or takes. 

What will we be doing: spawn `N` threads, each thread will send 10 requests. Measure time of each single request.

Every request is a single `HEAD` account.

## `HEAD`

```python
thread_number = [1, 2, 3, 4, 5, 7, 10, 20, 30, 40, 50, 75, 100]
```

With total request number increase it's becoming very difficult to test with single machine.

## Plotting

```python
for x in thread_number:
    plot('%dhead_%dthread_noconn.txt' % (x*10, x))
```

## Merging results

We vary requests intensity, so it _should_ somehow influence response time.

```python
x = thread_number
y = [stats('%dhead_%dthread_noconn.txt' % (d*10, d))[3] for d in x]

k, b = np.polyfit(x, y, 1)
x0 = np.linspace(0, 300, 10)
y0 = k*x0 + b
print 'time = %f * N + %f' % (k, b)
```

```python
_ = plt.scatter(x, y)
_ = plt.plot(x0, y0)
_ = plt.xlabel('threads')
_ = plt.ylabel('time, [s]')
```

Let's estimate response time standard deviation depending on thread count

```python
x = thread_number
y = [stats('%dhead_%dthread_noconn.txt' % (d*10, d))[4] for d in x]
```

```python
_ = plt.scatter(x, y)
_ = plt.xlabel('threads')
_ = plt.ylabel('time stddev, [s]')
```

# Conclusion

1. Median response time seems to have linear dependecy over thread count.
1. Standard deviation is big enough: this is the result of having "several" reponse times.
1. Can't vary thread count too much with only one machine.
1. Very difficult to test due to instability of results and lots of factors.
