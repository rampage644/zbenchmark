```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

Function `stats` will return 5-tuple (min, max, mean, median, std)

```python
def stats(file):
    with open(file, 'r') as ifile:
        l = [float(line) for line in ifile.readlines()]
    array = np.array(l)
    return (np.min(array),
            np.max(array),
            np.mean(array),
            np.median(array),
            np.std(array))
```

Function `plot` plots 2 histograms:
1. Original data
1. Filtered data

_Filtered_ data doesn't contain outliers. See http://en.wikipedia.org/wiki/68%E2%80%9395%E2%80%9399.7_rule

```python
def plot(file):
    with open(file, 'r') as ifile:
        l = [float(line) for line in ifile.readlines()]

    array = np.array(l)
    criteria = np.median(array) + 3 * np.std(array)
    array_filtered = np.array([el for el in l if el < criteria])
    plt.figure().set_size_inches(16, 5)
    plt.subplot(1,2,1)
    _ = plt.hist(array, bins=30)
    plt.subplot(1,2,2)
    _ = plt.hist(array_filtered, bins=30)
```
