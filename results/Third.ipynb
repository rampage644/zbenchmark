# Notebook setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

Function `stats` will return 5-tuple (min, max, mean, median, std)

```python
def stats(file):
    with open(file, 'r') as ifile:
        l = [float(line) for line in ifile.readlines()]
    array = np.array(l)
    return (np.min(array),
            np.max(array),
            np.mean(array),
            np.median(array),
            np.std(array))
```

Function `plot` plots 2 histograms:
1. Original data
1. Filtered data

_Filtered_ data doesn't contain outliers. See http://en.wikipedia.org/wiki/68%E2%80%9395%E2%80%9399.7_rule

```python
def plot(file):
    with open(file, 'r') as ifile:
        l = [float(line) for line in ifile.readlines()]

    array = np.array(l)
    criteria = np.median(array) + 3 * np.std(array)
    array_filtered = np.array([el for el in l if el < criteria])
    plt.figure().set_size_inches(16, 5)
    plt.title(file)
    plt.subplot(1,2,1)
    _ = plt.hist(array, bins=75)
    plt.subplot(1,2,2)
    _ = plt.hist(array_filtered, bins=75)
```

# Experimenting

First, we fire N requests (request = HEADing an account) in M threads. Fixing N = 500.
Will vary N from 10 to 250.

```python
thread_number = [10, 20, 30, 40, 50, 75, 100, 125, 150, 175, 200, 250]
```

## Plotting

```python
for x in thread_number:
    plot('500head_%dthread.txt' % x)
```

## Merging results

We vary fire intensity, so it _should_ somehow influence response time.

```python
x = thread_number
y = [stats('500head_%dthread.txt' % d)[3] for d in x]

k, b = np.polyfit(x, y, 1)
x0 = np.linspace(0, 300, 10)
y0 = k*x0 + b
print 'time = %f * N + %f' % (k, b)
```

```python
_ = plt.scatter(x, y)
_ = plt.plot(x0, y0)
_ = plt.xlabel('threads')
_ = plt.ylabel('time, [s]')
```

Let's estimate response time standard deviation depending on thread count

```python
x = thread_number
y = [stats('500head_%dthread.txt' % d)[4] for d in x]
```

```python
_ = plt.scatter(x, y)
_ = plt.xlabel('threads')
_ = plt.ylabel('time stddev, [s]')
```

# Conclusion

1. We still could get response time around 7s 
1. We could also get time around 12s, 40s, etc
1. Median response time increases as thread number increase 
1. Standard deviation also increases
